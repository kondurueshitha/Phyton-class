Distplots
Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.



```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([22,33,44,55])

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([22,33,44,55], hist=False) #The difference between the above grapgh and this is we are plotting graph with outhistogram

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head  # will show the data
```

```python
iris.describe()
```

```python
sns.set()

sns. swarmplot(x="species", y="petal_length", data=iris)
```

```python
sns.set()

sns.pairplot(iris, hue='species', size=2);
```

SCATTER PLOT USING SEABORN

```python
df = pd.read_csv(r"file.csv")
df.head() 
```

```python
sns.relplot(x="Random", y="Choice", data = df) #To draw the scatter plot, weâ€™ll be using the relplot() function of the seaborn library.
```

```python
sns.relplot(x="Random", y="Choice", hue = "Name", data = df)
```

Plotting Categorical Data
1.Jitter
2.Hue
3.Boxplot
4.Voilin Plot
5.Pointplot

```python

sns.catplot(x="Random", y="Choice", data=df)
```

```python

sns.catplot(x="Random", y="Choice", data=df,jitter = False) #Jitter is the deviation from the true value
```

Box plot

```python
sns.catplot(x="Random", y="Choice", data=df,kind = "box")
```

Violin 

```python
sns.catplot(x="Random", y="Choice", data=df,kind = "violin")
```

Point plot

```python
sns.catplot(x="Random", y="Choice", data=df,kind = "point")
```

Heatmaps using Seaborn

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(corrmat, vmax=.8)
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(5, 8))
sns.heatmap(corrmat, vmax=.5)
```
